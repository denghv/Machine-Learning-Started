# Functions in Python
Function is a block of code that can be reused in a Program

Factorial of a number is the product of the positive integers less than or equal to the given number

Factorial of 5 = 5 * 4 * 3 * 2 * 1 = 120

```python
number = int(input('Enter a number to find its factorial : '))

factorial = 1

if number == 0:
  print(' The Factorial is 1')
else:
  for i in range(1, number+1):  # range(1, 6) --> 1,2,3,4,5
    factorial = factorial * i
  print('The factorial of ', number,' is ', factorial)
```

Factorial Function

```python
def factorial_value(num):
  factorial = 1

  if num == 0:
    return factorial

  else:
    for i in range(1, num+1):
      factorial = factorial * i
    return factorial
```

```python
print(factorial_value(5))
```
